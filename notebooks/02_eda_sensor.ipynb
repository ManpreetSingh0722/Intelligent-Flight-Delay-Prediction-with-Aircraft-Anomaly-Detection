# EDA: Sensor Data Analysis

Exploratory Data Analysis for CMAPSS sensor dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load sensor data
df = pd.read_csv('../data/raw/cmapss_sensor_data.csv')
df.head()
```

```python
# Basic statistics
print(df.info())
print(df.describe())
```

```python
# Correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix - Sensor Data')
plt.show()
```

```python
# Sensor readings over time
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

axes[0, 0].plot(df['engine_temp'])
axes[0, 0].set_title('Engine Temperature')

axes[0, 1].plot(df['pressure'])
axes[0, 1].set_title('Pressure')

axes[1, 0].plot(df['vibration'])
axes[1, 0].set_title('Vibration')

axes[1, 1].plot(df['altitude'])
axes[1, 1].set_title('Altitude')

plt.tight_layout()
plt.show()
```
