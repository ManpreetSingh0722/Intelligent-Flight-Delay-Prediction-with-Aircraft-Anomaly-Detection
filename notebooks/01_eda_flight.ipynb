# EDA: Flight Data Analysis

Exploratory Data Analysis for flight delay dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load flight data
df = pd.read_csv('../data/raw/flight_data.csv')
df.head()
```

```python
# Basic statistics
print(df.info())
print(df.describe())
```

```python
# Missing values
print(df.isnull().sum())
```

```python
# Visualization: Delay distribution
plt.figure(figsize=(10, 6))
plt.hist(df['delay_minutes'], bins=30, edgecolor='black')
plt.xlabel('Delay (minutes)')
plt.ylabel('Frequency')
plt.title('Distribution of Flight Delays')
plt.show()
```
