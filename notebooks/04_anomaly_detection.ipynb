# Anomaly Detection in Sensor Data

Detecting anomalies in aircraft sensor readings using Isolation Forest

```python
import sys
sys.path.insert(0, '../src')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from anomaly_detection import detect_anomalies
from utils import save_model

# Load sensor data
df = pd.read_csv('../data/processed/sensor_processed.csv')
print(f"Data shape: {df.shape}")
df.head()
```

```python
# Perform anomaly detection
X = df.drop('unit_id', axis=1)
predictions, detector = detect_anomalies(X.values, contamination=0.05)

# Count anomalies
n_anomalies = (predictions == -1).sum()
n_normal = (predictions == 1).sum()

print(f"Normal samples: {n_normal}")
print(f"Anomalies detected: {n_anomalies}")
```

```python
# Visualize anomalies
anomaly_scores = detector.score(X.values)

plt.figure(figsize=(12, 6))
plt.scatter(range(len(anomaly_scores)), anomaly_scores, c=predictions, cmap='coolwarm')
plt.xlabel('Sample Index')
plt.ylabel('Anomaly Score')
plt.title('Anomaly Detection Scores')
plt.colorbar(label='Normal (1) / Anomaly (-1)')
plt.show()
```

```python
# Save the anomaly detector
save_model(detector, '../models/anomaly_model.pkl')
```
